# Input/Output with pySNOW

Here we show how to read / write data to standard data formats (xyz files and lammps dump/data files).

--------------

```python
from snow.io.xyz import read_xyz, read_xyz_movie
from snow.io.lammps import read_lammps_data, read_order_lammps_dump
```

### .xyz files

.xyz files are a very common format for storing atomistic data. For a given (list of) configurations, the data has the following structure:

`Natoms`

`comment line`

`chemical_symbol_atom1` `coord1_atom1` `coord2_atom1` `coord3_atom1`

`chemical_symbol_atom2` `coord1_atom2` `coord2_atom2` `coord3_atom2`


`...`

`Natoms`

`comment line`

`...`

Let's read xyz files with pySNOW:

```python
file_config   = 'tutorial_structures/Au13_ih.xyz'  #file containing data about a single configuration (a 13-atoms gold icosahedron)
file_trainset = 'tutorial_structures/train.xyz'    #a file with several configs with a varying number of atoms each (a training set for a machine-learning potential)

el_ico, coords_ico = read_xyz(file_config)

elements_list, coords_list = read_xyz_movie(file_trainset)

#iterate on the frames of your 'movie':
for el, coords in zip(elements_list, coords_list):
    #do something with the elements list
    #do something with the coordinates np.ndarray
    pass
```

Some xyz files have extra columns of per-atom data (e.g. forces acting on the i-th atom). To load such data, specify the list of ids of the columns you want to load data from (notice that i=0,1,2,3 are the chemical specie and positions - so you might want to start from column_id=4).

data from these columns is returned in a list of shape `(natoms, n_extra_cols)`.

```python
forces_cols_ids = [4,5,6]
el, coords, forces = read_xyz_movie(file_trainset, extra_cols_indexes=forces_cols_ids)
```

### Lammps-related formats

There are two lammps related formats we deal with: the `.data` format and the `lammps dump` format. 

Notice that by default lammps dump files do not follow a precise and consistent ordering of the atoms in the various frames. The function to load data in pysnow reorders automatically lammps dumps to have consistent ordering in your coordinates arrays, but you need to have told lammps to print the atomic ids in your lammps dump.

A further caveat: lammps does not dump atomic species by deafult, but reassigns an identifier to each chemical specie in the system. Thus, the element vector will actually contain the atomic specie id lammps printed, and you will need to map it back to the chemical element.

```python
lammps_data_file = 'tutorial_structures/Au13_ih.data'
lammps_dump_file = 'tutorial_structures/example.dump'

#read lammps dump format
el_list, coords_list = read_order_lammps_dump(lammps_dump_file)
print(el_list[0][0:10]) #chemical symbols are mapped to numbers

#read lammps data format
el, coords = read_lammps_data(lammps_data_file)
```
