# Tutorial - shape descriptors in SNOW

Here we show how to to use snow to compute the shape descriptors of nanoparticles. We will deal with descriptors obtained from the gyration tensor and inertia tensor, namely: gyration radius, asphericity, acilindricity, relative shape anisotropy, and aspect ratio. 

The gyration tensor is defined as
$$
S_{ij}=\frac{1}{N}\sum_{k=1}^Nr_{i,k}r_{j,k}
$$
where $r_{i,k}$ is the $i$-th coordinate of the $k$-th particle in a system of $N$ total particles. Given $\lambda_1<\lambda_2<\lambda_3$ the eigenvalues of the gyration tensor, we can define a series of descriptors of the geometric shape of the distribution of points (here, the atomic positions). These are:

 - the gyration radius $R_g=\lambda_1+\lambda_2+\lambda_3$, which gives information on the spatial extend of the distribution of atoms,
 - the asphericity $b=\lambda_3-(\lambda_2+\lambda_1)/2$, which is 0 for distributions that are symmetric with respect to the three coordinate axes - e.g. in spherical symmetry,
 - the acylindricity $b=\lambda_2-\lambda_1$, which is 0 for distributions that are symmetric with respect to two coordinate axes - e.g. in cylindrical symmetry,
 - the relative shape anisotropy $\kappa^2=\frac{3}{2}\frac{\lambda_1^2+\lambda_2^2+\lambda_3^2}{(\lambda_1 + \lambda_2 + \lambda_3)^2}-\frac{1}{2}$, which is 0 for spherically symmetric distributions, and 1 for linear distributions.

The inertia tensor is defined as
$$
I_{ij} = \sum_{k=1}^Nm_k(|\mathbf{r}_k|^2\delta_{ij}-r_{i,k}r_{j,k})
$$
where symbols are as above.

A further geometrical descriptor can be derived from here: the aspect ratio is defined as the ratio between the smallest and largest eigenvalues of the inertia tensor.
**********************

```python
#load some data
from snow.io.xyz import read_xyz

_, coords_ico = read_xyz('tutorial_structures/Cu309_Ih.xyz') #a 5-layer copper icosahedron
_, coords_dh  = read_xyz('tutorial_structures/Cu247_Dh.xyz') #a (3,15,0) copper decahedron
```

We now compute the descriptors for a geometrically built Icosahedron nanoparticle.

```python
from snow.descriptors.shape_descriptors import gyr_desc, aspect_ratio, gyr_rad

asph, acil, rsa = gyr_desc(coords_ico)
ar = aspect_ratio(coords_ico)
rgyr = gyr_rad(coords_ico)

print(' ------ computed descriptors ------- ')
print(f'      Gyration radius [Å]: {rgyr:.3f}')
print(f'              Asphericity: {asph:.3e}')
print(f'            Acilindricity: {acil:.3e}')
print(f'Relative shape anisotropy: {rsa:.3e}')
print(f'             Aspect ratio: {ar:.3e}')
print(' ----------------------------------- ')
```

A geometrically built Icosahedron is a very spherical and regular particle, thus, descriptors quantifying its deviation from spherical or cilyndrical symmetry (or its shape anisotropy) will fall close to zero, while the aspect ratio is exactly 1. The gyration radius gives a measure of the spatial extent of the nanoparticle.

_________________

We now compute descriptors for a very elongated, geometrically built Decahedron with $(m,n,p) = (3,15,0)$.

```python
asph, acil, rsa = gyr_desc(coords_dh)
ar = aspect_ratio(coords_dh)
rgyr = gyr_rad(coords_dh)

print(' ------ computed descriptors ------- ')
print(f'      Gyration radius [Å]: {rgyr:.3f}')
print(f'              Asphericity: {asph:.3e}')
print(f'            Acilindricity: {acil:.3e}')
print(f'Relative shape anisotropy: {rsa:.3e}')
print(f'             Aspect ratio: {ar:.3e}')
print(' ----------------------------------- ')
```

Here, asphericity, relative shape anisotropy, and aspect ratio all label a configuration which is significantly different from a sphere. A cilyndrical symmetry is retained, this explains the very small acilindricity value.
