# Tutorial - shape descriptors in SNOW

Here we show how to to use snow to compute the shape descriptors of nanoparticles. We will deal with descriptors obtained from the gyration tensor and inertia tensor, namely: gyration radius, asphericity, acilindricity, relative shape anisotropy, and aspect ratio. Check out our paper for an intepretation of these quantities.

**********************

```python
#load some data
from snow.io.xyz import read_xyz

_, coords_ico = read_xyz('tutorials_structures/Cu309_Ih.xyz') #a 5-layer copper icosahedron
_, coords_dh  = read_xyz('tutorials_structures/Cu247_Dh.xyz') #a (3,15,0) copper decahedron
```

We now compute the descriptors for a geometrically built Icosahedron nanoparticle.

```python
from snow.descriptors.shape_descriptors import gyr_desc, aspect_ratio, gyr_rad

conf_index = 0 #for reference

asph, acil, rsa = gyr_desc(conf_index, coords_ico)
ar = aspect_ratio(conf_index, coords_ico)
rgyr = gyr_rad(conf_index, coords_ico)

print(' ------ computed descriptors ------- ')
print(f'      Gyration radius [Å]: {rgyr:.3f}')
print(f'              Asphericity: {asph:.3e}')
print(f'            Acilindricity: {acil:.3e}')
print(f'Relative shape anisotropy: {rsa:.3e}')
print(f'             Aspect ratio: {ar:.3e}')
print(' ----------------------------------- ')
```

A geometrically built Icosahedron is a very spherical and regular particle, thus, descriptors quantifying its deviation from spherical or cilyndrical symmetry (or its shape anisotropy) will fall close to zero, while the aspect ratio is exactly 1. The gyration radius gives a measure of the spatial extent of the nanoparticle.

_________________

We now compute descriptors for a very elongated, geometrically built Decahedron with $(m,n,p) = (3,15,0)$.

```python
asph, acil, rsa = gyr_desc(conf_index, coords_dh)
ar = aspect_ratio(conf_index, coords_dh)
rgyr = gyr_rad(conf_index, coords_dh)

print(' ------ computed descriptors ------- ')
print(f'      Gyration radius [Å]: {rgyr:.3f}')
print(f'              Asphericity: {asph:.3e}')
print(f'            Acilindricity: {acil:.3e}')
print(f'Relative shape anisotropy: {rsa:.3e}')
print(f'             Aspect ratio: {ar:.3e}')
print(' ----------------------------------- ')
```

Here, asphericity, relative shape anisotropy, and aspect ratio all label a configuration which is significantly different from a sphere. A cilyndrical symmetry is retained, this explains the very small acilindricity value.
